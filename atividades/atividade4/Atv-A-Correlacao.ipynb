```python
import numpy as np

np.random.seed(42)

# Conjunto com relação linear
x1 = np.linspace(1, 9, 9)
y1 = 2 * x1 + np.random.normal(0, 1, 9)

# Conjunto aleatório uniforme (sem relação)
x2 = np.linspace(1, 9, 9)
y2 = np.random.uniform(1, 18, 9)
```

```python
from scipy.stats import linregress

# Calcular Regressão
reg1 = linregress(x1, y1)
reg2 = linregress(x2, y2)

# Calcular Covariância
cov1 = np.cov(x1, y1)[0, 1]
cov2 = np.cov(x2, y2)[0, 1]
```

```python
import matplotlib.pyplot as plt

def plotar_regressao(x, y, reg, cov, titulo):
    plt.figure(figsize=(6, 4))
    plt.scatter(x, y, label='Dados')
    plt.plot(x, reg.slope * x + reg.intercept, 'r-', label='Regressão')
    plt.title(titulo)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.show()
    
    print(f"{titulo}")
    print(f"  Correlação (r): {reg.rvalue:.2f}")
    print(f"  p-valor: {reg.pvalue:.4f}")
    print(f"  Covariância: {cov:.2f}")
    print('-'*50)

plotar_regressao(x1, y1, reg1, cov1, "Conjunto com correlação")
plotar_regressao(x2, y2, reg2, cov2, "Conjunto aleatório (uniforme)")
```


# Interpretaçao dos Resultados para os Dados com Correlaçao

## Correlação (r = 0.99):
Valor muito próximo de 1 → indica uma forte relação linear positiva.

Isso significa que à medida que x aumenta, y também aumenta de forma muito consistente. Os pontos estariam quase alinhados na reta.

## P-Value = 0.0000:
Abaixo de 0.05, significa que rejeitamos a hipótese nula (de que não há relação), a chance dessa relação ter ocorrido por acaso é praticamente zero.

Conclusão: a relação entre x e y é estatisticamente significativa.

## Covariância = 14.87:
Valor positivo e alto → indica que x e y crescem juntos.

# Interpretaçao dos Resultados para os Dados sem Correlaçao

## Correlação (r = 0.19):
Um valor próximo de 0, indica que não existe relação linear clara entre as variáveis.

## P-Value = 0.6165:
Valor muito maior que 0.05 → não rejeitamos a hipótese nula. Não há evidência estatística de relação entre x e y.

## Covariância = 1.47:
Valor positivo baixo, mostra que não há variação conjunta significativa.
