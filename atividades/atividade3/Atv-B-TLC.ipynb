```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import gamma
```

# Gerando uma população com distribuição gamma

```python
np.random.seed(42)
populacao = gamma.rvs(a=2.0, scale=2.0, size=10000)
```

```python
tamanhos_amostra = [10, 30, 50, 100]
n_amostras = 1000
```

# Histograma da distribuiçao da população

```python
plt.figure(figsize=(14, 6))
plt.subplot(2, 3, 1)
sns.histplot(populacao, bins=30, kde=True, color='skyblue')
plt.title("População (Distribuição Gama)")
plt.xlabel("Valores")
plt.ylabel("Frequência")
```

# Aplicaçao do Teorema do Limite Central

```python
for i, n in enumerate(tamanhos_amostra):
    medias_amostrais = []

    for _ in range(n_amostras):
        amostra = np.random.choice(populacao, size=n, replace=False)
        medias_amostrais.append(np.mean(amostra))

        plt.subplot(2, 3, i + 2)
    sns.histplot(medias_amostrais, bins=30, kde=True, color='orange')
    plt.title(f'Médias Amostrais (n={n})')
    plt.xlabel("Média da Amostra")
    plt.ylabel("Densidade")
    
plt.tight_layout()
plt.suptitle("Aplicação do TLC à Distribuição Gama", fontsize=16, y=1.03)
plt.show()
```

# Avaliaçao das amostras

# Amostra tamanho 10: 
A distribuição ainda está assimétrica, com cauda à direita.
Com apenas 10 elementos por amostra, a média amostral não é suficientemente para se aproximar de uma distribuiçao normal. 

# Amostra tamanho 30: 
A curva começa a assumir uma forma mais simétrica e centrada.
Com amostra de 30 elementos a curva começa a assumir uma forma mais simétrica e proxima de uma distribuiçao normal.

# Amostras de tamanho 50 e 100: 
A curva começa a assumir uma forma mais simétrica e centrada.
A distribuição das médias é bem proximo da distribuiçao Normal, a dispersão em torno da média diminui, e o formato da curva é quase simétrico.

# Conclusão: 
Pequenas amostras não são suficientes para que o TLC produza uma distribuição aproximadamente normal, quantos mais amostras mais simetrico a curva fica e mais proximo de uma distribuiçao nornal, atestanto o conceito do teorema do limite central.
