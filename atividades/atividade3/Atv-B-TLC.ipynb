```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import gamma
```

# Gerando uma população com distribuição gamma

```python
np.random.seed(42)
populacao = gamma.rvs(a=2.0, scale=2.0, size=10000)

# Tamanhos de amostra a serem testados
tamanhos_amostra = [10, 30, 50, 100]
n_amostras = 1000
```

# Histograma da distribuiçao da população

```python
plt.figure(figsize=(14, 6))
plt.subplot(2, 3, 1)
sns.histplot(populacao, bins=40, kde=True, color='skyblue')
plt.title("População (Distribuição Gama)")
plt.xlabel("Valores")
plt.ylabel("Frequência")
```

# Aplicaçao do TLC para cada amostra

```python
for i, n in enumerate(tamanhos_amostra):
    medias_amostrais = []

    for _ in range(n_amostras):
        amostra = np.random.choice(populacao, size=n, replace=False)
        medias_amostrais.append(np.mean(amostra))

        plt.subplot(2, 3, i + 2)
    sns.histplot(medias_amostrais, bins=30, kde=True, color='orange')
    plt.title(f'Médias Amostrais (n={n})')
    plt.xlabel("Média da Amostra")
    plt.ylabel("Densidade")
    
plt.tight_layout()
plt.suptitle("Aplicação do TLC à Distribuição Gama", fontsize=16, y=1.03)
plt.show()
```
